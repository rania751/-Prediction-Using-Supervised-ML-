#Importing libraries

```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

#Importing Data

```python
url = 'http://bit.ly/w-data'
scores = pd.read_csv(url)
print("the uploading is finished:  ")
scores.head()
```

#Visualising the Data

```python
#Ploting the distribution of scores
scores.plot(x='Hours', y='Scores', style='o', figsize=(12,7))
plt.title('Hours (H) and scores (S)')
plt.show()
```

```python
x = scores.iloc[:,:-1].values
y = scores.iloc[:, 1].values
```

#Training the model

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=30,test_size=0.3)
Regressor = LinearRegression()
Regressor.fit(x_train.reshape(-1,1), y_train)
```

#Plotting the regression line

```python
line = Regressor.coef_*x+Regressor.intercept_

#Plotting the test data
import seaborn as sea
sea.regplot(x=scores['Hours'], y=scores['Scores'],color='red')
```

#Predictions

```python
y_pred = Regressor.predict(x_test)
```

```python
#Comparing the original and the predicted data
df = pd.DataFrame({'original_scores': y_test, 'Predicted_scores': y_pred})
df
```

```python
#checking for given exemple in the task
hours =9.25
pred = Regressor.predict(np.array(hours).reshape(1,-1))
print("prediction =",pred[0])
```

```python
df.plot(kind='bar')
plt.show()
```

#The model evaluation

```python
print('MAE:',metrics.mean_absolute_error(y_test,y_pred))
print('MSE:',metrics.mean_squared_error(y_test,y_pred))
print(metrics.r2_score(y_test,y_pred))
```
